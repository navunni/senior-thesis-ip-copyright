```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from correlation_analysis import reports_dir 
```

```python
global_corr_dir = reports_dir / "global_correlation"
path = global_corr_dir / "spotify_popularity_correlation_summary_spotify_streams.csv"

df = pd.read_csv(path)

cat_order = [
    "remake_licensed",
    "copy_unauthorized",
    "ai_cover",
    "ai_melody",
    "inspired",
    "sampling",
    "remix_official",
    "remix_user",
]
df = df.set_index("event_category").loc[cat_order].reset_index()

x = np.arange(len(df))
width = 0.35

fig, ax = plt.subplots(figsize=(8, 4))

ax.bar(x - width/2, df["avg_pair_corr"], width, label="Pair Correlation")
ax.bar(x + width/2, df["avg_random_corr"], width, label="Random Baseline")

ax.set_xticks(x)
ax.set_xticklabels(df["event_category"], rotation=30, ha="right")
ax.set_ylabel("Average Pearson correlation")
ax.set_title("Average Originalâ€“Derived Correlations by Event Category\n(Spotify Streams, Weekly Changes)")
ax.legend()

plt.tight_layout()

out_path = global_corr_dir / "spotify_popularity_event_category_correlations.png"
plt.savefig(out_path, dpi=300)
plt.close()

print(f"Saved Figure 6 -> {out_path}")
```
